```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:\\Users\\mawat\\Downloads\\Book\\housing.csv")
data.head()
```

```python
corr = data.corr(method='pearson')
corr.head()
```

```python
plt.subplots(figsize=(11, 11))
plt.title('Pearson Correlation of features')
sns.heatmap(corr, annot=True, linewidth=.5, square=True)
```

The top variables in terms of correlating to median house value are:
- median_income

```python
mask = np.zeros_like(corr, dtype=np.bool)
mask[np.triu_indices_from(mask)] = True 
cmap = sns.diverging_palette(240, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, linewidths=.5, cmap=cmap, center=0)
```
