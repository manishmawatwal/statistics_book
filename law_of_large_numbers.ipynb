```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
```

```python
# creating a dice
dice = pd.Series([1, 2, 3, 4, 5, 6])
dice.to_frame()
```

```python
print('Avg of our dice:', dice.mean())
```

```python
def monte_carlo(series_list, n):
    result = []
    for i in range(1, n):
        result.append(series_list.sample(i, replace=True).mean())
    pd.Series(result).plot()
    plt.show()
    print('Average of our dice is:', sum(result)/len(result))
```

```python
monte_carlo(dice, 5)
```

```python
monte_carlo(dice, 10)
```

```python
monte_carlo(dice, 100)
```

```python
monte_carlo(dice, 1000)
```

```python
# result becomes closer to the expected value as the number of trials increase
monte_carlo(dice, 10000)
```
