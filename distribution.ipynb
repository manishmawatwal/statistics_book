```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)
inliers = np.random.normal(loc=10, scale=2, size=50)
outliers = np.random.normal(loc=30, scale=5, size=10)

# Create combined data array
data = np.concatenate((inliers, outliers))

# Create a figure and axis
plt.figure(figsize=(8, 6))
plt.title("Outliers and Inliers")
plt.xlabel("X")
plt.ylabel("Y")

# Plot inliers and outliers
plt.scatter(range(len(data)), data, c='blue', label='Inliers')
plt.scatter(range(len(inliers), len(data)), outliers, c='red', label='Outliers')

plt.legend()
plt.show()
```

```python
# Generate random inlier data
np.random.seed(42)
inliers_x = np.random.uniform(low=0, high=10, size=50)
inliers_y = 2 * inliers_x + np.random.normal(0, 1, size=50)

# Calculate the regression line parameters (slope and intercept)
slope, intercept = np.polyfit(inliers_x, inliers_y, 1)

# Create a figure and axis
plt.figure(figsize=(8, 6))
plt.title("Regression Line for Inliers")
plt.xlabel("X")
plt.ylabel("Y")

# Plot inliers
plt.scatter(inliers_x, inliers_y, c='blue', label='Inliers')

# Plot the regression line
regression_line_x = np.linspace(0, 10, 100)
regression_line_y = slope * regression_line_x + intercept
plt.plot(regression_line_x, regression_line_y, c='red', label='Regression Line')

plt.legend()
plt.show()
```

```python
# Generate random inlier and outlier data
np.random.seed(42)
inliers_x = np.random.uniform(low=0, high=10, size=50)
inliers_y = 2 * inliers_x + np.random.normal(0, 1, size=50)

outliers_x = np.random.uniform(low=0, high=10, size=10)
outliers_y = np.random.uniform(low=5, high=25, size=10)

# Calculate the regression line parameters (slope and intercept) for inliers
slope, intercept = np.polyfit(inliers_x, inliers_y, 1)

# Create a figure and axis
plt.figure(figsize=(8, 6))
plt.title("Regression Line with Inliers and Outliers")
plt.xlabel("X")
plt.ylabel("Y")

# Plot inliers and outliers
plt.scatter(inliers_x, inliers_y, c='blue', label='Inliers')
plt.scatter(outliers_x, outliers_y, c='red', label='Outliers')

# Plot the regression line
regression_line_x = np.linspace(0, 10, 100)
regression_line_y = slope * regression_line_x + intercept
plt.plot(regression_line_x, regression_line_y, c='green', label='Regression Line')

plt.legend()
plt.show()
```

```python
from scipy.stats import binom

# Parameters for the Binomial distribution
n = 10  # Number of trials
p = 0.3  # Probability of success

# Generate x values (number of successes)
x = np.arange(0, n+1)

# Calculate the PMF using scipy's binom.pmf function
pmf = binom.pmf(x, n, p)

# Create a bar plot of the PMF
plt.bar(x, pmf, align='center', alpha=0.7)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution PMF')
plt.xticks(x)
plt.show()
```

```python
from scipy.stats import norm

# Generate random data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=100)

# Sort the data in ascending order
sorted_data = np.sort(data)

# Create an array of probabilities for the cumulative distribution
probabilities = np.arange(1, len(sorted_data) + 1) / len(sorted_data)

# Plot the cumulative distribution
plt.plot(sorted_data, probabilities, marker='o', linestyle='none')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('Cumulative Distribution')
plt.grid(True)
plt.show()
```

```python
# Generate random data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=100)

# Sort the data in ascending order
sorted_data = np.sort(data)

# Create an array of probabilities for the cumulative distribution
probabilities = np.arange(1, len(sorted_data) + 1) / len(sorted_data)

# Plot the cumulative distribution using step function
plt.step(sorted_data, probabilities, where='mid')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('Smooth Cumulative Distribution')
plt.grid(True)
plt.show()
```

```python
# Generate random data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=100)

# Calculate mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data)

# Sort the data in ascending order
sorted_data = np.sort(data)

# Create an array of probabilities for the cumulative distribution
probabilities = np.arange(1, len(sorted_data) + 1) / len(sorted_data)

# Plot the cumulative distribution using step function
plt.step(sorted_data, probabilities, where='mid', label='CDF')
plt.axvline(mean, color='red', linestyle='dashed', linewidth=1.5, label='Mean')
plt.axvline(mean + std_dev, color='green', linestyle='dashed', linewidth=1.5, label='Mean + Std Dev')
plt.axvline(mean - std_dev, color='green', linestyle='dashed', linewidth=1.5, label='Mean - Std Dev')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.title('Smooth Cumulative Distribution with Mean and Std Dev')
plt.legend()
plt.grid(True)
plt.show()
```
